```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Dates":pd.to_datetime(["2022-01-01","2022-01-01","2022-02-02","2022-03-03","2022-04-04","2022-05-05","2022-06-06",]),
    "Value":[10,15,22,28,35,42,48]
}
```

```python
df=pd.DataFrame(data)
df.set_index("Dates",inplace=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker='X')
plt.xlabel("Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Month": ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"],
    "Sales": [110, 160, 140, 172, 190, 210, 235, 215, 205, 230, 235, 240]
}
```

```python
df = pd.DataFrame(data)
df.set_index("Month", inplace=True)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index, df["Sales"], marker="o")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=69)
plt.title("Time Series with Seasonality")
plt.grid(True)
plt.show()
```
