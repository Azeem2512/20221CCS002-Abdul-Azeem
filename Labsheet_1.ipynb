```python
import matplotlib.pyplot as plt
x = [1,2,3]
y = [4,5,5]
plt.plot(x,y)
plt.title("INFO")
plt.xlabel ("X axis")
plt.ylabel ("Y axis")
plt.show()
```

```python
x = [1,2,3,4]
y = [2,3,4,5]
mycolors = ["#FFFF00","#FFA500","#808000","#A52A2A"]
plt.bar(x,y,color=mycolors)
plt.title("INFO")
plt.xlabel ("X axis" ,fontsize = 20)
plt.ylabel ("Y axis",fontsize = 20)
plt.show()
```

```python
 y =[35,35,20,20,25,25]
 mylabels = ["Apples" , "Mango" , "Orange" , "Grapes","melon","berry"]
 mycolors = ["r","#FFFF00","#FFA500", "#00FF00","#808000","#A52A2A"]
 myexplode = [0.2,0.1,0,0.2,0,0]
 plt.pie (y, labels = mylabels , colors = mycolors , explode = myexplode ,  shadow=True )
 plt.show()
```

```python
x = [1,2,3,4]
y = [2,3,4,5]
mycolors = ["#FFFF00","#FFA500","#808000","#A52A2A"]
plt.barh(x,y,color=mycolors)
plt.title("INFO")
plt.xlabel ("X axis" ,fontsize = 20)
plt.ylabel ("Y axis",fontsize = 20)
plt.show()
```

```python
x = [1,2,3,4]
y = [2,3,4,5]
siz = [ 80, 150, 200, 300]
mycolors = ["#FFFF00","#FFA500","#808000","#A52A2A"]
plt.scatter(x,y,color=mycolors,marker ="*" , edgecolors='g', s= siz , alpha=1 , linewidths= 1 ,cmap='viridis' )
plt.title("INFO")
plt.xlabel ("X axis" )
plt.ylabel ("Y axis")
plt.show()

```
